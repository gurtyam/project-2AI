```python
%%writefile data/stories.yml

version: "3.1"

stories:
- story: greet_and_inquire_budget
  steps:
  - intent: greet
  - action: utter_greet
  - intent: budget_status
  - action: utter_budget_status

- story: add_expense_and_check_balance
  steps:
  - intent: add_expense
  - action: utter_ask_expense_details
  - intent: provide_expense_details
  - action: action_add_expense
  - action: utter_expense_added
  - intent: budget_status
  - action: utter_budget_status

- story: inquire_income
  steps:
  - intent: income
  - action: utter_income

- story: set_savings_goal
  steps:
  - intent: savings_goal
  - action: utter_ask_savings_goal
  - intent: provide_savings_goal
  - action: action_set_savings_goal
  - action: utter_savings_goal

- story: inquire_investment_options
  steps:
  - intent: investment_options
  - action: utter_investment_options

- story: end_conversation
  steps:
  - intent: goodbye
  - action: utter_goodbye
```

```python
%%writefile data/nlu.yml

version: "3.1"

nlu:
- intent: greet
  examples: |
    - hi
    - hello
    - hey

- intent: budget_status
  examples: |
    - What is my current budget status?
    - Can you check how much money I have left in my budget?
    - Tell me about my budget balance.

- intent: add_expense
  examples: |
    - Record an expense of $50 for lunch.
    - Add $30 for groceries to my expenses.
    - Log a spending of $100 for utilities.

- intent: income
  examples: |
    - How much did I earn this month?
    - Tell me about my income.
    - What is my total salary?

- intent: savings_goal
  examples: |
    - Set a savings goal of $500 for this month.
    - How can I save money more effectively?
    - Advise on achieving my financial goals.

- intent: investment_options
  examples: |
    - What are the investment options available?
    - Tell me about potential investment opportunities.
    - Guide me on smart investments.

- intent: goodbye
  examples: |
    - Bye
    - Goodbye
    - See you later

- intent: provide_expense_details
  examples: |
    - I spent $50 on lunch today.
    - Groceries cost me $30.
    - Utilities expense was $100.

- intent: provide_savings_goal
  examples: |
    - I want to save $500 this month.
    - Set a savings goal of $800 for me.

```

```python
%%writefile domain.yml

version: "3.1"

intents:
  - greet
  - goodbye
  - budget_status
  - add_expense
  - income
  - savings_goal
  - investment_options
  - provide_expense_details
  - provide_savings_goal

responses:
  utter_greet:
    - text: "Hello! How can I assist you with your personal finances today?"

  utter_budget_status:
    - text: "Your current budget balance is $XXX."

  utter_ask_expense_details:
    - text: "Sure! Please provide details about the expense, including the amount and description."

  utter_expense_added:
    - text: "Expense of $XXX has been recorded. Your updated budget balance is $YYY."

  utter_income:
    - text: "Your total income for this month is $XXX."

  utter_ask_savings_goal:
    - text: "Great! What is your savings goal for this month?"

  utter_savings_goal:
    - text: "Awesome! Your savings goal of $XXX has been set for this month."

  utter_investment_options:
    - text: "Here are some investment options you can explore: [Option 1], [Option 2], [Option 3]."

  utter_goodbye:
    - text: "Goodbye! If you have more questions, feel free to ask."

  utter_iamabot:
    - text: "I am a bot, powered by Rasa."

actions:
  - action_add_expense
  - action_set_savings_goal
```

```python
# Train a ML model
model_path = rasa.train(domain, config, [training_files], output)
print(model_path)
```

```python
from rasa.jupyter import chat
#endpoints = 'endpoints.yml'

endpoints = None
```

```python
# Chat with your assistant#
# To start chatting to an assistant, call the chat function, passing in the path to your saved model.
# If you do not have custom actions you can set endpoints = None or omit it:

chat(model_path.model, endpoints)
```

```python
# Evaluate your model against test data#
# Rasa has a convenience function for getting your training data.
# Rasa's get_core_directory and get_nlu_directory are functions which recursively find all the stories or NLU data files and
# copies them into temporary directories.
# The return values are the paths to these newly created directories.

import rasa.shared.data as data
nlu_data_directory = data.get_nlu_directory(training_files)
stories_directory = data.get_core_directory(training_files)
print(stories_directory, nlu_data_directory)
rasa.test(model_path, stories_directory, nlu_data_directory)
print("Done testing.")
```
